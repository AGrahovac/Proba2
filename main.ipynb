```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
cD = pd.read_csv('countries.csv')
def country(countryName):
    return cD[cD.country == countryName]
afg = country("Afghanistan")
plt.plot(afg.year, afg.population/afg.population.iloc[0])
US = country("United States")
plt.plot(US.year, US.population/US.population.iloc[0])
plt.title("Afghanistan/US")
plt.xlabel("Year")
plt.ylabel("Population growth")
plt.legend(['Afghanistan', 'US'])
plt.show()
```

```python
def country2(cN):
    x = cD[cD.country == cN]
    plt.plot(x.year, x.population/x.population.iloc[0],label=cN)
    
 
country2("Afghanistan")
country2("United States")
country2("Serbia")
country2("Croatia")
country2("Ireland")

plt.title("Growth in time")
plt.xlabel("Year")
plt.ylabel("Population growth")
plt.legend()
plt.show()
```
